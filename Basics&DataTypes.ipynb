**Type Function:** type() functions returns the type of the specified object.

```python
print(type(12))
print(type(12.5))
print(type(True))
print(type('1.0'))
print(type(-0))
print(type(0.0))
print(type('hello'))
```

# **Typecasting:**
 Typecasting is the process of converting one datatype into another.


```python
# int- str, bool, float

print(int(3.14), int(True), int(False), int('3'))
#print(int('3.5')) ---> will throw an error
#print(int('a'))   ---> will throw an error

# str to -> int, bool, float
print(str(3.14), str(True), str(False), str(3))

# float to -> int, str, bool
print(float(3), float(True), float(False), float('3'))

# boolto -> int, str, float
print(bool(3), bool(True), bool(False), bool('3'))
print(bool('0'),bool(0),bool(0.0), bool('False'), bool(''), bool(' '))
```
