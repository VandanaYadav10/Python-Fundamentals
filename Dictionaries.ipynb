# ***Dictionaries:***
Dictionaries are built-in data structures that stores values in Key:Value pairs. The values in a dictionary can be of any data type and can be duplicated, whereas Keys can't be duplicated and must be immutable. A dictinary and be created using curly brackets({}).

```python
dct= {1:'Tanish',2:'Manish',3:'Rajnish'}        # values are added in pairs(key:value)
dct[4]='Mohit'                                  # to add a new pair to the dictionary
dct
```

```python
len(dct)                                        # each pair is an elt of a dictionary
```

```python
data= {1:{'Name':'Tanish','Phone':8734587345,'Address':'Noida','Marks':{'Hindi':45,'English':54,'Math':56}},
      2:{'Name':'Manish','Phone':8734587346,'Address':'Noida','Marks':{'Hindi':67,'English':78,'Math':90}},
      3:{'Name':'Nitesh','Phone':8734587387,'Address':'Noida','Marks':{'Hindi':35,'English':53,'Math':34}}}

for Roll_No in data.keys(): #would take only the key values skipping the ones that are not there
  Total_Marks=sum(list(data[Roll_No]['Marks'].values())) #converts the values of the dic into a list then sums it.
  Total_exams=len(data[Roll_No]['Marks'].values())
  print('Roll No.   : '+str(Roll_No))
  print('Name.      : '+str(data[Roll_No]['Name']))
  print('Total Marks: '+str(Total_Marks))
  print('Average    : '+str(round((Total_Marks/Total_exams),2)))
  print('-'*25)
```

```python
data= {1:{'Name':'Ashish','Phone':8734587345,'Address':'Noida','Marks':{'Hindi':45,'English':54,'Math':56}},
      2:{'Name':'Manish','Phone':8734587346,'Address':'Noida','Marks':{'Hindi':67,'English':78,'Math':90}},
      3:{'Name':'Nitesh','Phone':8734587387,'Address':'Noida','Marks':{'Hindi':35,'English':53,'Math':34}}}

for roll_no in range (1,4):
  Total_Marks=sum(list(data[roll_no]['Marks'].values()))
  Total_exams=len(data[roll_no]['Marks'].values())
  data[roll_no]['Average']=round((Total_Marks/Total_exams),2)

print(data)
```
