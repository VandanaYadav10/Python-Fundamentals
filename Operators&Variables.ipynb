# **Operators**
Operators in general are used to perform operations on values and variables.



## **Arithmetic Operators :**
Arithmetic operators in Python are used to perform common mathematical operations such as addition, subtraction, multiplication and division.


```python
## Arithmetic operators on integers
print(2+3)     # Addition
print(2-3)     # Subtraction
print(2*3)     # Multiplication
print(2/3)     # Division
print(4/2)     # Always returns a float value
print(2//3)    # Floor Division
print(2%3)     # Modulus
print(2**3)    # Exponent
```

```python
## Arithmetic operators on Strings and Booleans
print(3*'Python')       # Reproduces the text n number of times
print('Python'+'Class') # Concatenates
print(True+True+True)
print(True+False+False)
print('Python'*True)
print('Python'*False)
print(5+True)
print(5*True)

```

## **Comparison Operators :**
Comparison operators in Python compare values and return a Boolean result.


```python
print(2==3) # equals to
print(2!=3) # not equals to
print(2>3)  # Greater than
print(2<3)  # Lesser than
print(2>=3) # Greater than equal to
print(2<=3) # Lesser than equal to
```

##**Assignment of values to variables :**




```python
a=10       # Variable 'a' has been assigned a value 10
b=20       # Variable 'b' has been assigned a value 20

#Arithmetic Operations on the Variables defined above:
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```
