# ***For Loop:***
A For Loop in Python is used to iterate over a sequence (list, tuple, string) or other iterable objects.

```python
for i in range(10):        #This parameter refers to the range which in this case is 0-9
  print(i)
```

```python
for i in range(1,11):       #The first parameter is the starting point, second is the range.
  print(i)
```

***Print the table of 13.***

```python
n=13
for i in range(1,11):
  print(n,'*',i,'=',n*i)
```

***Print the table of a number n.***

```python
n = 35

for i in range(n,n*10+1,n):     #the third parameter is the step size.
  print(i)
```

***Table of 13 in reverse order:***

```python
n=13

for i in range(n*10,n-1,-n):
  print(i)
```

***Write a code to print the sum of first n numbers.***

```python
n=7
s=0
for i in range(1,n+1):
  s=s+i
print(s)
```

# ***Nested For Loop:***

***Two dices are rolled. What is the probability of getting a sum of eight?***

```python
c=0
num=8

for i in range(1,7):
 for j in range(1,7):         #nested for loop
  if i+j==num:
    c+=1
print(round((c/36)*100,2),'%')
```

***Three dices are rolled. What is gthe probability of getting the sum as nine?***

```python
c=0
num=9
for i in range (1,7):
  for j in range (1,7):
    for k in range (1,7):
      if i+j+k==num:
        c+=1
print(round((c/216)*100,2),'%')
```
