## ***IF Else Statements:***
If-Else is a fundamental control flow structure used for decision making.
These statements allows execution of specific blocks of codes depending on the condition being True or not.

***Write a program to check whether the user is eligible for voting or not.***

If the age belongs to the following intervals, display the corresponding message as output:
*   (-inf,0) or (100,inf):- not a valid age
*   '[0,17]:- Not Eligible

*   '[18,100]:- Eligible


```python
age= -1
if((age>=0)and (age<=17)):
  print('not eligible')
elif((age>=18)and(age<100)):
  print('eligible')
else:
    print('not valid age')
```

***Write a program to check if the given number is divisible by 5 and 7 or not, if it is divisible then square it other wise return 'not divisible' as an answer.***

```python
n=35
if((n%5==0) and (n%7==0)):
  print(n**2)
else:
  print('not divisible')
```
